# Create Datasets & DataLoaders

```python
import datasets
from datasets import Dataset
from transformers import DataCollatorWithPadding
from transformers import TrainingArguments
from transformers import Trainer
```

```python
# Split the dataset into train and test sets
train_df, val_df = train_test_split(trdf1, test_size=0.2, random_state=42)

train_df['label'] = train_df['label'].apply(lambda x: int(x))
train_df['text'] = train_df['text'].apply(lambda x: str(x))

val_df['label'] = val_df['label'].apply(lambda x: int(x))
val_df['text'] = val_df['text'].apply(lambda x: str(x))

tedf1['label'] = tedf1['label'].apply(lambda x: int(x))
tedf1['text'] = tedf1['text'].apply(lambda x: str(x))

train_dataset =  Dataset.from_dict({"data": train_df["text"], "label": train_df["label"]}).with_format("torch")
val_dataset =  Dataset.from_dict({"data": val_df["text"], "label": val_df["label"]}).with_format("torch")
test_dataset =  Dataset.from_dict({"data": tedf1["text"], "label": tedf1["label"]}).with_format("torch")

print(len(train_dataset))
print(len(val_dataset))
print(len(test_dataset))
```

```python
def tokenize_function(example):
    return tokenizer(example["data"], truncation=True)


tokenized_train_dataset = train_dataset.map(tokenize_function, batched=True)
tokenized_val_dataset = val_dataset.map(tokenize_function, batched=True)
tokenized_test_dataset = test_dataset.map(tokenize_function, batched=True)

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

# Train Model

```python
from sklearn.metrics import accuracy_score, f1_score

# Define evaluation metrics function
def compute_metrics(p):
    predictions = p.predictions.argmax(axis=1)
    labels = p.label_ids

    accuracy = accuracy_score(labels, predictions)
    f1 = f1_score(labels, predictions, average='binary')


    # return {"accuracy": accuracy, "f1": f1, "predictions": list(predictions)}
    return {"accuracy": accuracy, "f1": f1}
```

```python
batch_size = 16

training_args = TrainingArguments("train_logs", evaluation_strategy="epoch",\
                                  learning_rate=2e-6,
                                  num_train_epochs=10,
                                  lr_scheduler_type="cosine",
                                  optim = "adamw_torch",
                                  weight_decay=0,
                                  per_device_train_batch_size=batch_size,
                                  per_device_eval_batch_size=batch_size,
                                  run_name= "bert-base-multilingual-cased", # "roberta-base-hate-best",  # name of the W&B run (optional)
                                  logging_steps=1,  # how often to log to W&B
                                  )


trainer = Trainer(
    model,
    training_args,
    train_dataset=tokenized_train_dataset,
    eval_dataset=tokenized_val_dataset,
    data_collator=data_collator,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
)
```

```python
trainer.train()
```

#### Model Comparison on Dev Set (Our Test Set)

```python
# Calculate accuracy for each model
def calculate_accuracy(df, column_name):
  if column_name == 'our_prediction':
    correct_predictions = df[df['our_model_prediction'] == df['label']].shape[0]
  elif column_name == 'uncased_prediction':
    correct_predictions = df[df['uncase_model_prediction'] == df['label']].shape[0]
  elif column_name == 'cased_prediction':
    correct_predictions = df[df['case_model_prediction'] == df['label']].shape[0]
  else:
    correct_predictions = df[df['RoBERT_prediction'] == df['label']].shape[0]
  total_predictions = df.shape[0]
  accuracy = correct_predictions / total_predictions
  return accuracy

our_accuracy = calculate_accuracy(filtered_df, 'our_prediction')
uncased_accuracy = calculate_accuracy(filtered_df, 'uncased_prediction')
cased_accuracy = calculate_accuracy(filtered_df, 'cased_prediction')
RoBERT_accuracy = calculate_accuracy(filtered_df, 'RoBERT_prediction')

# Print out the accuracies
print(f"Our Model's Accuracy: {our_accuracy}")
print(f"Uncased Model's Accuracy: {uncased_accuracy}")
print(f"Cased Model's Accuracy: {cased_accuracy}")
print(f"RoBERT Model's Accuracy: {RoBERT_accuracy}")
```
